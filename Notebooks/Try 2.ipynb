```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style


```

```python
data = pd.read_csv('test_datasetOG.csv')
print(data.shape)
data.head()
```

```python
x= data['expiry_in_months'].values
y=data['sellingprice_based_on_exp_in_month'].values
print(x.shape)
print(y.shape,'y')
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)

n= len(x)

num=0
den=0
for i in range(n):
    num +=(x[i]-mean_x)*(y[i]-mean_y)
    den += (x[i]-mean_x)**2
b1= num/den
b0=mean_y-(b1*mean_x)

print(b1,b0)
```
